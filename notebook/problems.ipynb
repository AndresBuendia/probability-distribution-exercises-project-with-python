## Binomial distribution

In probability theory and statistics, the binomial distribution is the discrete probability distribution that gives only two possible results in an experiment, for example, when you flip a coin it can by either head or tails.

```python
from scipy.stats import binom

# Define all the probability functions related to this distribution

def dbinom(x, size, prob = 0.5):
    """
    Calculates the point estimate of the binomial distribution
    """
    result = binom.pmf(k = x, n = size, p = prob, loc = 0)

    return result

def pbinom(q, size, prob = 0.5):
    """
    Calculates the cumulative of the binomial distribution
    """
    result = binom.cdf(k = q, n = size, p = prob, loc = 0)

    return result

def qbinom(p, size, prob = 0.5):
    """
    Calculates the quantile function from the binomial distribution
    """
    result = binom.ppf(q = p, n = size, p = prob, loc = 0)

    return result

def rbinom(n, size, prob = 0.5):
    """
    Generates random variables from the binomial distribution
    """
    result = binom.rvs(n = size, p = prob, size = n)

    return result
```

```python
# Binomial(10, 0.2) distribution where n = 10 is the number of trials and p = 0.2 is the probability of success in a single trial.
import numpy as np
np.random.seed(42)

print(f"Probability that a Binomial(10, 0.2) takes the value 2: {dbinom(2, size = 10, prob = 0.2)}")
print(f"Probability that a Binomial(10, 0.2) will take a value less than 2: {pbinom(2, size = 10, prob = 0.2)}")
print(f"Which value of a Binomial(10, 0.2) has a cumulative probability of 0.9? {qbinom(0.9, size = 10, prob = 0.2)}")
print(f"Generate 2000 random values from a Binomial(10, 0.2) distribution: {rbinom(2000, size = 10, prob = 0.2)}")
```

```python
import matplotlib.pyplot as plt

np.random.seed(42)

size = 10
prob = 0.2
data = rbinom(2000, size = 10, prob = 0.2)

pmf = dbinom(data, size = size, prob = prob)
cdf = pbinom(data, size = size, prob = prob)
ppf = qbinom(data, size = size, prob = prob)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].stem(data, pmf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
from scipy.stats import norm, uniform
import numpy as np
import matplotlib.pyplot as plt

def dnormal(x, loc=0, scale=1):
    result = norm.pdf(x, loc, scale)
    return result
def pnormal(x, loc=0, scale=1):
    result = norm.cdf(x, loc, scale)
    return result
def qnormal(x, loc=0, scale=1):
    result = norm.ppf(x, loc, scale)
    return result
def rnormal(loc=0, scale=1, size=1, random_state=None):
    result = norm.rvs(loc, scale, size,)
    return result
def duniform(x, loc=0, scale=1):
    result = uniform.pdf(x, loc, scale)
    return result
def puniform(x, loc=0, scale=1):
    result = uniform.cdf(x, loc, scale)
    return result
def quniform(x, loc=0, scale=1):
    result = uniform.ppf(x, loc, scale)
    return result
def runiform(loc=0, scale=1, size=1):
    result = uniform.rvs(loc, scale, size)
    return result

print(f"Probability that a Normal(0, 1) takes the value 2: {dnormal(2, scale = 1, loc = 0)}")
print(f"Probability that a Normal(0, 1) will take a value less than 2: {pnormal(2, scale = 1, loc = 0)}")
print(f"Which value of a Normal(0, 1) has a cumulative probability of 0.9? {qnormal(0.9, scale = 1, loc = 0)}")
print(f"Generate 2000 random values from a Normal(0, 1) distribution: {rnormal(size = 2000, scale = 1, loc = 0)}")


print(f"Probability that a Uniform(0, 1) takes the value 2: {duniform(2, scale = 1, loc = 0)}")
print(f"Probability that a Uniform(0, 1) will take a value less than 2: {puniform(2, scale = 1, loc = 0)}")
print(f"Which value of a Uniform(0, 1) has a cumulative probability of 0.9? {quniform(0.9, scale = 1, loc = 0)}")
print(f"Generate 2000 random values from a Uniform(0, 1) distribution: {runiform(size = 2000, scale = 1, loc = 0)}")


sd = 1
mean = 0
tam = 100

datan = rnormal(size= tam, scale = sd, loc = mean)
datau = runiform(size = tam, scale = sd, loc = mean)

pdfn = dnormal(datan, scale = sd, loc = mean)
cdfn = pnormal(datan, scale = sd, loc = mean)
ppfn = qnormal(datan, scale = sd, loc = mean)

pdfu = duniform(datau, scale = sd, loc = mean)
cdfu = puniform(datau, scale = sd, loc = mean)
ppfu = quniform(datau, scale = sd, loc = mean)

fig, axis = plt.subplots(3, 2, figsize = (10, 14))

axis[0,0].stem(datan, pdfn, "r-")
axis[1,0].stem(datan, cdfn, "b-")
axis[2,0].stem(datan, ppfn, "g-")

axis[0, 0].set_xlabel('Datos')
axis[0, 0].set_ylabel('PDF')
axis[0, 0].set_title('Distribución Normal - PDF')

axis[1, 0].set_xlabel('Datos')
axis[1, 0].set_ylabel('CDF')
axis[1, 0].set_title('Distribución Normal - CDF')

axis[2, 0].set_xlabel('Datos')
axis[2, 0].set_ylabel('PPF')
axis[2, 0].set_title('Distribución Normal - PPF')

axis[0,1].stem(datau, pdfu, "r-")
axis[1,1].stem(datau, cdfu, "b-")
axis[2,1].stem(datau, ppfu, "g-")

axis[0, 1].set_xlabel('Datos')
axis[0, 1].set_ylabel('PDF')
axis[0, 1].set_title('Distribución Uniforme - PDF')

axis[1, 1].set_xlabel('Datos')
axis[1, 1].set_ylabel('CDF')
axis[1, 1].set_title('Distribución Uniforme - CDF')

axis[2, 1].set_xlabel('Datos')
axis[2, 1].set_ylabel('PPF')
axis[2, 1].set_title('Distribución Uniforme - PPF')

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```
